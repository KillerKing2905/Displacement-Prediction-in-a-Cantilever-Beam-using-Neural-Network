```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models

# Define function to load and preprocess displacement data from file
def load_displacement_file(file_path):
    # Load displacement data from file
    displacement_data = np.loadtxt(file_path)  # Load all columns
    # Preprocess data as needed (e.g., normalization)
    displacement_data = (displacement_data - np.mean(displacement_data)) / np.std(displacement_data)  # Normalization and standardization
    return displacement_data

# Define function to load and preprocess prop data from file
def load_prop_file(file_path):
    # Define which columns to load
    columns_to_load = (0,)

    # Load data using np.loadtxt with usecols parameter
    data = np.loadtxt(file_path, usecols=columns_to_load)

    return data

from tensorflow.keras import layers, models

def create_cnn_model(sequence_length, num_features, num_classes):
    model = models.Sequential([
        layers.Conv1D(32, 3, activation='relu', input_shape=(sequence_length, num_features)),
        layers.MaxPooling1D(2),
        layers.Conv1D(64, 3, activation='relu'),
        layers.MaxPooling1D(2),
        layers.Conv1D(64, 3, activation='relu'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model


# Define function to train CNN model
def train_cnn_model(model, X_train, y_train, X_val, y_val, epochs):
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(X_train, y_train, epochs=epochs, validation_data=(X_val, y_val))
    return history

# Load displacement files and preprocess
displacement_files = [f"/content/drive/MyDrive/femcode11(DL_executed)-20240513T044442Z-001/femcode(DL_executed)/cont_ver2/displacement{i}" for i in range(1, 400)]
displacement_data = []

for file_path in displacement_files:
    data = load_displacement_file(file_path)
    displacement_data.append(data)

# Convert list to numpy array
displacement_data = np.array(displacement_data)

# Load load file
load_file_path = "/content/drive/MyDrive/femcode11(DL_executed)-20240513T044442Z-001/femcode(DL_executed)/cont_ver2/load.dat"
load_data = load_displacement_file(load_file_path)

# Load prop files and preprocess
prop_files = [f"/content/drive/MyDrive/femcode11(DL_executed)-20240513T044442Z-001/femcode(DL_executed)/cont_ver2/prop{i}.dat" for i in range(1, 500)]
prop_data = []

for file_path in prop_files:
    data = load_prop_file(file_path)
    prop_data.append(data)

# Convert list to numpy array
prop_data = np.array(prop_data)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(prop_data, displacement_data, test_size=0.2, random_state=42)

# Create CNN model
model = create_cnn_model(19, 19, 1, 1681 * 3)  # Input shape based on the dimensions of your prop files and displacement files

# Train CNN model
history = train_cnn_model(model, X_train, y_train, X_val, y_val, epochs=50)

# Predict displacement
predictions = model.predict(X_val)  # Assuming X_val contains prop data for prediction







import numpy as np

def load_prop_file(file_path):
    # Define which columns to load
    columns_to_load = (0, 1)  # Example: load only the first two columns

    # Load data using np.loadtxt with usecols parameter
    data = np.loadtxt(file_path, usecols=columns_to_load)

    return data

# Example usage
prop_files = ['file1.txt', 'file2.txt', 'file3.txt']  # Example list of file paths

prop_data = []
for file_path in prop_files:
    data = load_prop_file(file_path)
    prop_data.append(data)

# Now prop_data contains the loaded data from each file
```
